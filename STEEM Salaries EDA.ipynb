<h3>importing neccessary libraries</h3>

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid", color_codes=True)
sns.set_palette('crest_r', 30)
print('Importing libraries finished')
```

<h3>Loading data from csv file</h3>

```python
data = pd.read_csv('Levels_Fyi_Salary_Data.csv')
df = data.copy()
print('Data has been seccuessfully loaded and was coppied to new varriable')
```

```python
df.head()
```

<h2><center>Cleaning Data</center></h2>

```python
df[['company','level','gender','otherdetails', 'Education', 'Race', 'tag']] = df[['company','level','gender','otherdetails', 'Education', 'Race', 'tag']].fillna('unknown')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.dropna(how='all', inplace=True)
print('NaN values have been droped or replaced properly')
```

```python
df.head()
```

```python
def find_education_degree(row):
    if row[17] == 1:
        return 'Doctorate_Degree'
    elif row[15] == 1:
        return 'Masters_Degree'
    elif row[16] == 1:
        return 'Bachelors_Degree'
    elif row[19] == 1:
        return 'Some_College'
    elif row[18] == 1:
        return 'Highschool'
    else:
        return 'unknown'    

education = df.apply(find_education_degree, axis='columns')
df['education'] = education
print('Education column was added based on the highest degree of the employee')
```

```python
columns_to_drop = ['Doctorate_Degree','Masters_Degree', 'Bachelors_Degree', 'Some_College', 'Highschool' ]
df.drop(columns_to_drop, axis=1, inplace=True)
df.drop('rowNumber',axis=1, inplace=True)
df.drop(df.columns[16:21], axis=1, inplace=True)
df['gender'] = df['gender'].replace('Title: Senior Software Engineer', 'unknown')
print('Unusable columns were droped')
```

```python
def calculate_total_income(row):
    total_sal = max([row['totalyearlycompensation'], row['basesalary']])
    if total_sal == 0:
        return 0
    else:
        return total_sal + row['bonus'] + row['stockgrantvalue']

total_income = df.apply(calculate_total_income, axis=1)
df.insert(12, 'total income', total_income)
print('Total income was calculated and added to the dataset')
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)
print('White spaces were removed from the dataset')
```

### Maximun percentage of job from total in company

```python
company_job_count_per_job_title = df.groupby(['company', 'title']).agg({'title': 'count'})
company_emp_size_per_jon_title = df.groupby('company')['title'].count()
company_job_count_per_job_title = company_job_count_per_job_title.merge(company_emp_size_per_jon_title,
                                                                        left_index=True,
                                                                       right_index=True).reset_index()
company_job_count_per_job_title.rename(columns={'title_x': 'job title count', 'title_y':'total'}, inplace=True)
company_job_count_per_job_title['ratio'] = company_job_count_per_job_title['job title count'] / company_job_count_per_job_title['total']
company_job_count_per_job_title_filt = company_job_count_per_job_title['total'] > 100
company_job_count_per_job_title = company_job_count_per_job_title[company_job_count_per_job_title_filt]
company_job_count_per_job_title = company_job_count_per_job_title.groupby('title').apply(lambda x: x.iloc[x['ratio'].argmax()]).sort_values('ratio', ascending=False)
company_job_count_per_job_title['ratio'] = (company_job_count_per_job_title['ratio'].round(2) * 100)
```

```python
plt.figure(figsize=(18,7))
splot = sns.barplot(x=company_job_count_per_job_title['title'], y= company_job_count_per_job_title['ratio'])

for i in range(len(splot.patches)):
    splot.annotate(company_job_count_per_job_title.iloc[i,0], 
                   (splot.patches[i].get_x() + splot.patches[i].get_width() / 2., splot.patches[i].get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 35),
                   textcoords = 'offset points', fontsize=16, rotation=75, color='grey')
plt.margins(y=0.3)
plt.xticks(rotation=75, fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Job title', fontsize=16)
plt.ylabel('Ratio in %', fontsize=16)
plt.title('Highest job title count per total employees ratio for each job title', fontsize=20)
plt.show()
```
